```python
def getValueDistribution(col):
    distr = {}
    for cell in col:
        for token in cell.split():
            if token not in distr:
                distr[token] = 1
            else:
                distr[token] = distr[token]+1
    return distr
```

```python
def plotHorizontalBarChart(dictionary,attribute):
    plt.rcdefaults()
    fig, ax = plt.subplots()

    # Example data
    y_pos = np.arange(len(list(dictionary.keys())))

    ax.barh(list(dictionary.keys()),list(dictionary.values()), align='center')
    ax.set_yticks(y_pos)
    ax.set_yticklabels(list(dictionary.keys()))
    ax.invert_yaxis()  # labels read top-to-bottom
    ax.set_xlabel('Occurrences')
    ax.set_title(attribute)
    plt.show()
```
