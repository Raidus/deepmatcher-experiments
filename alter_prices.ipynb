```python
test_df = pd.read_csv(os.path.join(data_dir,'test.csv'))
```

```python
altered_df = test_df[0:40].copy()
```

```python
altered_df.head(10)
```

```python
import random
def generate_random_price():
    dollars= str(random.randint(0,1000))
    cents = random.randint(0,100)
    if cents<=9:
        cents = str('0'+str(cents))
    else:
        cents = str(cents)
    return '$'+' '+dollars+'.'+cents
```

```python
print(generate_random_price())
print(generate_random_price())
```

```python
for i in range(0,10):
    altered_df.loc[i,'left_Price'] = generate_random_price()
    altered_df.loc[i,'right_Price'] = generate_random_price()
for i in range(10,20):
    r_price = generate_random_price()
    altered_df.loc[i,'left_Price'] = r_price
    altered_df.loc[i,'right_Price'] = r_price
for i in range(20,30):
    altered_df.loc[i,'right_Price'] = generate_random_price()
for i in range(30,40):
    altered_df.loc[i,'left_Price'] = generate_random_price()
```

```python
altered_df.head(20)
```

```python
altered_df.to_csv('sample_data/itunes-amazon/price-test.csv',index=False)
```
