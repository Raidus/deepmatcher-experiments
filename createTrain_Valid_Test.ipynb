```python
dir_path ='sample_data/itunes-amazon/split-files'
```

```python
import os
import pandas as pd
df_tableA = pd.read_csv(os.path.join(dir_path,'tableA.csv'))
df_tableB = pd.read_csv(os.path.join(dir_path,'tableB.csv'))
df_train = pd.read_csv(os.path.join(dir_path,'new_training_ids.csv'))
df_valid = pd.read_csv(os.path.join(dir_path,'validation_ids.csv'))
df_test = pd.read_csv(os.path.join(dir_path,'test_ids.csv'))
```

```python
df_tableA.head()
```

```python
df_tableB.head()
```

```python
df_tableA.columns=['id','left_Song_Name','left_Artist_Name','left_Album_Name','left_Genre','left_Price','left_CopyRight','left_Time',
                 'left_Released']
df_tableA.head()
```

```python
df_tableB.columns=['id','right_Song_Name','right_Artist_Name','right_Album_Name','right_Genre','right_Price','right_CopyRight',
                   'right_Time','right_Released']
df_tableB.head()
```

```python
#P sta per parziale
ptrain=pd.merge(df_train,df_tableA, how='inner', left_on='ltable_id', right_on='id')
pvalid=pd.merge(df_valid,df_tableA,how='inner',left_on='ltable_id',right_on='id')
ptest=pd.merge(df_test,df_tableA,how='inner',left_on='ltable_id',right_on='id')
```

```python
train=pd.merge(ptrain,df_tableB,how='inner',left_on='rtable_id',right_on='id')
valid=pd.merge(pvalid,df_tableB,how='inner',left_on='rtable_id',right_on='id')
test=pd.merge(ptest,df_tableB,how='inner',left_on='rtable_id',right_on='id')
```

```python
import random
train['id'] = (train['ltable_id']+ train['rtable_id']-1000)
train.tail()
```

```python
train = train.drop(['ltable_id','rtable_id','id_x','id_y'],axis=1)
```

```python
valid =valid.drop(['ltable_id','rtable_id','id_x','id_y'],axis=1)
test = test.drop(['ltable_id','rtable_id','id_x','id_y'],axis=1)
```
