# Define method to uniform rankings

```python
def compareRankings(r1,r2,leftPrefix='ltable_',rightPrefix='rtable_'):
    old_col = list(r1)
    for col in list(r2):
        r1[col] =  (r1[leftPrefix+col]+r1[rightPrefix+col])/2
    r1 = r1.drop(columns=old_col,axis=1)
    r1_sum = r1.sum()
    max_val = max(r1_sum)
    r1_sum_normalized = (r1_sum/max_val)
    r2_sum = r2.sum()
    max_val_r2 = max(r2_sum)
    r2_sum_normalized = (r2_sum/max_val_r2)
    r2_sum_normalized = r2_sum_normalized.apply(lambda i:1-i)
    return r1_sum_normalized,r2_sum_normalized
```

## Beer advo

```python
r1_neg = pd.read_csv('attacksResults/testneg_beer.csv')
r2_neg = pd.read_csv('notebooks/beerAdvo/experiment-results/testneg_ri_norms.csv')
r1_norm_neg,r2_norm_neg = compareRankings(r1_neg,r2_neg)
```

```python
r3 = pd.DataFrame(r1_norm_neg)
r4 = pd.DataFrame(r2_norm_neg)
pd.concat([r3,r4],axis=1)
```

```python
r1_norm_neg.plot.bar()
```

```python
r2_norm_neg.plot.bar()
```

```python
r1_pos = pd.read_csv('attacksResults/testpos_beer.csv')
r2_pos = pd.read_csv('notebooks/beerAdvo/experiment-results/testpos_ri_norms.csv')
r1_norm,r2_norm = compareRankings(r1_pos,r2_pos)
r1_norm.plot.bar()
```
