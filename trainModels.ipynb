```python
import deepmatcher as dm
train,validation,test = dm.data.process('sample_data/itunes-amazon/',train='train.csv',validation='validation.csv',
                                       test='test.csv',left_prefix='ltable_',right_prefix='rtable_')
```

```python
hybrid1 = dm.MatchingModel(attr_summarizer='hybrid')
```

```python
hybrid1.load_state('models/hybrid1.pth')
```

```python
hybrid1.run_eval(test)
```

```python
hybrid1.run_train(train,validation,best_save_path='models/hybrid1.pth',pos_neg_ratio=4,epochs=20,batch_size=16)
```

```python
hybrid2 = dm.MatchingModel(attr_summarizer='hybrid')
```

```python
hybrid2.run_train(train,validation,best_save_path='models/hybrid2.pth',pos_neg_ratio=4,epochs=20,batch_size=16)
```

```python
hybrid1.run_eval(test)
hybrid2.run_eval(test)
```
