```python
import random
def generate_random_time():
    minutes = str(random.randint(1,10))
    seconds = random.randint(0,60)
    if seconds<=9:
        seconds= '0'+str(seconds)
    else:
        seconds = str(seconds)
    return minutes+':'+seconds
```

```python
t1 = generate_random_time()
t1
```

```python
altered_positives_samples_df = pd.read_csv('sample_data/itunes-amazon/altered_positives_samples.csv')
altered_positives_samples_df.head(10)
```

```python
altered_positives_samples_df.shape
```

```python
altered_positives_samples_df['left_Time'] = altered_positives_samples_df['left_Time'].apply(
    lambda t: generate_random_time())
```

```python
altered_positives_samples_df['right_Time'] = altered_positives_samples_df['right_Time'].apply(
    lambda t: generate_random_time())
```
